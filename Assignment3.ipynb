```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import data, wb
import datetime
%matplotlib inline
```

# Read the stock information for Amazon (Tick: AMZN) using Pandas’ data reader

```python
#Define Start/End dates, and ticker names
start = datetime.datetime(2022, 1, 1)
end = datetime.datetime(2022, 2, 15)
```

```python
Amazon= data.DataReader("AMZN", 'yahoo', start, end)
```

# Amazon Table

```python
Amazon_ = Amazon.reset_index()
Amazon_
```

# Plot a time series (line chart) of the “volume” from January 1, 2022 to today.


```python
plt.style.use('ggplot')
Amazon_.plot.line(x='Date',y='Volume',figsize=(15,5),lw=2,title="Volume values from Jan.1 2022 to Feb.15 2022")
plt.show()
```

# Plot a bar chart of the “opening “ for the month of December 2021

```python
start2 = datetime.datetime(2021, 12, 1)
end2 = datetime.datetime(2021, 12, 31)
Amazondec= data.DataReader("AMZN", 'yahoo', start2, end2)
Amazondec_ = Amazondec.reset_index()
```

```python
plt.style.use('ggplot')
Amazondec_.plot.bar(x='Date',y='Open',figsize=(15,5),title="Opening values for December 2021", color="#770000")
plt.show()
```
